```python
def player_statistics(player_id):
    player_metadata = players.find_player_by_id(player_id)
    df = playercareerstats.PlayerCareerStats(player_id).get_data_frames()[0]
    player_df = df.copy()
    
    player_df['PPG'] = player_df['PTS'] / player_df['GP']
    player_df['PLAYER_NAME'] = players.find_player_by_id(player_id)['full_name']
    
    return player_df

```
